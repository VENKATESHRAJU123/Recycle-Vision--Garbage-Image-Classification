```python
# Cell 1: Import All Required Libraries
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Add src to path
sys.path.append('../src')

# Import custom modules
from preprocessing import split_dataset, create_data_generators
from config import *

print("✓ All libraries imported successfully!")
print(f"NumPy version: {np.__version__}")

# Cell 2: Create All Required Directories
print("\n" + "="*60)
print("CREATING PROJECT DIRECTORIES")
print("="*60)

# Directories to create
required_dirs = [
    '../outputs',
    '../outputs/visualizations',
    '../outputs/reports',
    '../outputs/logs',
    '../models',
    '../models/checkpoints',
    '../models/metadata'
]

for dir_path in required_dirs:
    os.makedirs(dir_path, exist_ok=True)
    if os.path.exists(dir_path):
        print(f"  ✓ {dir_path}")
    else:
        print(f"  ❌ Failed to create: {dir_path}")

print("\n✓ All directories ready!")

# Cell 3: Helper Function for Safe Saving
def safe_save_figure(filename, dpi=300, bbox_inches='tight'):
    """Safely save figure with automatic directory creation"""
    directory = os.path.dirname(filename)
    if directory and not os.path.exists(directory):
        os.makedirs(directory, exist_ok=True)
    plt.savefig(filename, dpi=dpi, bbox_inches=bbox_inches)
    return filename

def safe_save_csv(df, filename):
    """Safely save CSV with automatic directory creation"""
    directory = os.path.dirname(filename)
    if directory and not os.path.exists(directory):
        os.makedirs(directory, exist_ok=True)
    df.to_csv(filename)
    return filename

print("✓ Helper functions loaded!")

# Cell 4: Split Dataset
print("\n" + "="*60)
print("SPLITTING DATASET")
print("="*60)

split_dataset(
    source_dir=RAW_DATA_DIR, 
    dest_dir=PROCESSED_DATA_DIR, 
    test_size=0.15,
    val_size=0.15,
    random_state=42
)

print("\n✓ Dataset split completed!")

# Cell 5: Create Data Generators
print("\n" + "="*60)
print("CREATING DATA GENERATORS")
print("="*60)

train_generator, val_generator, test_generator = create_data_generators()

print("\n✓ Data generators created successfully!")
print(f"\nDataset Summary:")
print(f"  Training samples:   {train_generator.samples}")
print(f"  Validation samples: {val_generator.samples}")
print(f"  Test samples:       {test_generator.samples}")
print(f"  Total samples:      {train_generator.samples + val_generator.samples + test_generator.samples}")

# Cell 6: Visualize Augmented Images
print("\n" + "="*60)
print("VISUALIZING AUGMENTED IMAGES")
print("="*60)

# Get a batch of images
x_batch, y_batch = next(train_generator)

# Display augmented images
fig, axes = plt.subplots(4, 4, figsize=(15, 15))
fig.suptitle('Augmented Training Images (After Data Augmentation)', 
             fontsize=16, fontweight='bold', y=0.98)

for i, ax in enumerate(axes.flat):
    if i < len(x_batch):
        ax.imshow(x_batch[i])
        class_idx = np.argmax(y_batch[i])
        class_name = CLASS_NAMES[class_idx]
        ax.set_title(f'{class_name}', fontsize=12, fontweight='bold')
        ax.axis('off')

plt.tight_layout()
saved_path = safe_save_figure('../outputs/visualizations/augmented_images.png')
plt.show()

print(f"✓ Saved to: {saved_path}")

# Cell 7: Compare Original vs Augmented
print("\n" + "="*60)
print("COMPARING ORIGINAL VS AUGMENTED IMAGES")
print("="*60)

from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Generator without augmentation
original_datagen = ImageDataGenerator(rescale=1./255)
original_generator = original_datagen.flow_from_directory(
    os.path.join(PROCESSED_DATA_DIR, 'train'),
    target_size=IMG_SIZE,
    batch_size=1,
    class_mode='categorical',
    shuffle=True
)

# Get one original image
x_original, y_original = next(original_generator)

# Create augmented versions
augmented_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=ROTATION_RANGE,
    width_shift_range=WIDTH_SHIFT_RANGE,
    height_shift_range=HEIGHT_SHIFT_RANGE,
    shear_range=SHEAR_RANGE,
    zoom_range=ZOOM_RANGE,
    horizontal_flip=HORIZONTAL_FLIP,
    fill_mode=FILL_MODE
)

class_idx = np.argmax(y_original[0])
class_name = CLASS_NAMES[class_idx]

# Generate visualization
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle(f'Original vs Augmented: {class_name}', 
             fontsize=16, fontweight='bold')

# Show original
axes[0, 0].imshow(x_original[0])
axes[0, 0].set_title('ORIGINAL', fontsize=12, fontweight='bold', color='green')
axes[0, 0].axis('off')

# Generate augmented versions
augmented_generator = augmented_datagen.flow(x_original, batch_size=1, shuffle=False)

for i in range(7):
    row = (i + 1) // 4
    col = (i + 1) % 4
    augmented_img = next(augmented_generator)[0]
    axes[row, col].imshow(augmented_img)
    axes[row, col].set_title(f'Augmented {i+1}', fontsize=10)
    axes[row, col].axis('off')

plt.tight_layout()
saved_path = safe_save_figure('../outputs/visualizations/original_vs_augmented.png')
plt.show()

print(f"✓ Saved to: {saved_path}")

# Cell 8: Verify Data Distribution
print("\n" + "="*60)
print("VERIFYING DATA DISTRIBUTION")
print("="*60)

import pandas as pd

distribution_data = []

for split in ['train', 'validation', 'test']:
    split_path = os.path.join(PROCESSED_DATA_DIR, split)
    
    for class_name in CLASS_NAMES:
        class_path = os.path.join(split_path, class_name)
        
        if os.path.exists(class_path):
            num_images = len([f for f in os.listdir(class_path) 
                            if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
            
            distribution_data.append({
                'Split': split,
                'Class': class_name,
                'Count': num_images
            })

df_distribution = pd.DataFrame(distribution_data)
df_pivot = df_distribution.pivot(index='Class', columns='Split', values='Count')
df_pivot['Total'] = df_pivot.sum(axis=1)

print("\nData Distribution Table:")
print("="*60)
print(df_pivot)
print("="*60)

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart
df_pivot[['train', 'validation', 'test']].plot(
    kind='bar', 
    ax=axes[0],
    color=['#3498db', '#2ecc71', '#e74c3c'],
    edgecolor='black',
    linewidth=1.5
)
axes[0].set_title('Data Distribution Across Splits', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Class', fontsize=12)
axes[0].set_ylabel('Number of Images', fontsize=12)
axes[0].legend(title='Split', fontsize=10)
axes[0].grid(axis='y', alpha=0.3)
plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45, ha='right')

# Percentage chart
df_percent = df_pivot[['train', 'validation', 'test']].div(df_pivot['Total'], axis=0) * 100
df_percent.plot(
    kind='bar', 
    stacked=True,
    ax=axes[1],
    color=['#3498db', '#2ecc71', '#e74c3c'],
    edgecolor='black',
    linewidth=1.5
)
axes[1].set_title('Percentage Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Class', fontsize=12)
axes[1].set_ylabel('Percentage (%)', fontsize=12)
axes[1].legend(title='Split', fontsize=10)
axes[1].grid(axis='y', alpha=0.3)
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45, ha='right')

plt.tight_layout()
saved_path = safe_save_figure('../outputs/visualizations/data_distribution.png')
plt.show()

# Save CSV
csv_path = safe_save_csv(df_pivot, '../outputs/reports/data_distribution.csv')
print(f"\n✓ Saved distribution to: {csv_path}")

# Cell 9: Sample Images from Each Split
print("\n" + "="*60)
print("SAMPLING IMAGES FROM EACH SPLIT")
print("="*60)

def show_split_samples(split_name, num_samples=3):
    """Show sample images from a specific split"""
    split_path = os.path.join(PROCESSED_DATA_DIR, split_name)
    
    fig, axes = plt.subplots(len(CLASS_NAMES), num_samples, 
                            figsize=(num_samples*4, len(CLASS_NAMES)*3))
    fig.suptitle(f'Sample Images - {split_name.upper()} SET', 
                 fontsize=16, fontweight='bold', y=0.995)
    
    for i, class_name in enumerate(CLASS_NAMES):
        class_path = os.path.join(split_path, class_name)
        images = [f for f in os.listdir(class_path) 
                 if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:num_samples]
        
        for j, img_name in enumerate(images):
            img_path = os.path.join(class_path, img_name)
            img = Image.open(img_path)
            
            ax = axes[i, j] if len(CLASS_NAMES) > 1 else axes[j]
            ax.imshow(img)
            ax.axis('off')
            
            if j == 0:
                ax.set_ylabel(class_name, fontsize=12, fontweight='bold', 
                            rotation=0, ha='right', va='center')
    
    plt.tight_layout()
    saved_path = safe_save_figure(f'../outputs/visualizations/{split_name}_samples.png')
    plt.show()
    return saved_path

# Show samples
for split in ['train', 'validation', 'test']:
    path = show_split_samples(split, num_samples=3)
    print(f"✓ {split.capitalize()} samples saved to: {path}")

# Cell 10: Final Summary
print("\n" + "="*60)
print("PREPROCESSING SUMMARY REPORT")
print("="*60)

summary = {
    'Total Images': train_generator.samples + val_generator.samples + test_generator.samples,
    'Training Images': train_generator.samples,
    'Validation Images': val_generator.samples,
    'Test Images': test_generator.samples,
    'Number of Classes': NUM_CLASSES,
    'Image Size': f'{IMG_HEIGHT}x{IMG_WIDTH}',
    'Batch Size': BATCH_SIZE,
}

print("\n📊 Configuration:")
for key, value in summary.items():
    print(f"  {key:25} → {value}")

print("\n" + "="*60)
print("✓ PREPROCESSING COMPLETED SUCCESSFULLY!")
print("="*60)

print("\n📁 Generated Files:")
files = [
    'outputs/visualizations/augmented_images.png',
    'outputs/visualizations/original_vs_augmented.png',
    'outputs/visualizations/data_distribution.png',
    'outputs/visualizations/train_samples.png',
    'outputs/visualizations/validation_samples.png',
    'outputs/visualizations/test_samples.png',
    'outputs/reports/data_distribution.csv'
]

for f in files:
    full_path = os.path.join('..', f)
    if os.path.exists(full_path):
        print(f"  ✓ {f}")
    else:
        print(f"  ❌ Missing: {f}")

print("\n🚀 Next Step: Run 04_model_training_baseline.ipynb")
```
