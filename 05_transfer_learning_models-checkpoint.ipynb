```python
# Cell 1: Import libraries
import sys
sys.path.append('../src')

from train import train_all_models

# Cell 2: Train all models
results = train_all_models()

# Cell 3: Display results
import pandas as pd
import matplotlib.pyplot as plt

df_results = pd.DataFrame(results).T
df_results = df_results.sort_values('val_accuracy', ascending=False)

print("Model Comparison:")
print(df_results)

# Plot comparison
fig, ax = plt.subplots(figsize=(12, 6))
df_results['val_accuracy'].plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
ax.set_title('Model Comparison - Validation Accuracy', fontsize=16, fontweight='bold')
ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Validation Accuracy', fontsize=12)
ax.set_ylim([0, 1])
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(df_results['val_accuracy']):
    ax.text(i, v + 0.01, f'{v:.4f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../outputs/visualizations/model_comparison.png', dpi=300)
plt.show()

print("âœ“ All models trained and compared!")
```
